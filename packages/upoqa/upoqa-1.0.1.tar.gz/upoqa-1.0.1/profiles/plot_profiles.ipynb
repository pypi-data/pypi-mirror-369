```python
import matplotlib.pyplot as plt
from upoqa.problems import *
from profiles import plot_performance_profile_noise_free, plot_data_profile_noise_free, plot_speedup_profile_noise_free
from utils import *
from solvers import *
plt.style.use('default')
loaded_info = problem_set.load_problem_set("s2mpj_prob_set_v1")

problem_idx_range = problem_set.valid_problem_idx()
if problem_set.enabled_problem_set == "s2mpj_prob_set_v1":
    black_list = [
        54, # "MINSURF"  # all solvers start from 1.0000000000000007 and end with 1.0000000000000007, meaningless
    ]
    for idx in black_list:
        problem_idx_range.remove(idx)
```

# Speed-up Profile

```python
solver_identifier_and_names = [
    (None, "bobyqa"),
    (None, "l-bfgs-b (ffd)"),
    (None, "upoqa_(single)"),
    (None, "upoqa"),
]

ftols = [1e-1, 1e-4, 1e-7]
ftolstr = [-1, -4, -7]
fig_size = (6, 3.35)
plot_fig_and_ax = None
for i in range(len(ftols)):
    ftol = ftols[i]
    _, figg, axx = plot_speedup_profile_noise_free(
        solver_labels = [x[1] for x in solver_identifier_and_names],
        solver_identifiers = [x[0] for x in solver_identifier_and_names],
        problem_idx_range = problem_idx_range,
        plot_label = r"$\epsilon = 10^{" + f"{ftolstr[i]}" + r"}$",
        plot_fig_and_ax = plot_fig_and_ax,
        ftol = ftol,
        fig_size = fig_size,
        # fmt = 'r--'
        # grid_on = False,
    )
    plot_fig_and_ax = (figg, axx)
```
