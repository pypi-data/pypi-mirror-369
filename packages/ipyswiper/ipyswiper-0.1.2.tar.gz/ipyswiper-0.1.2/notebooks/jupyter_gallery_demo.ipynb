```python
from ipyswiper import IpySwiper
```

```python
images = [
    {"label": "Image 1", "image": 'https://picsum.photos/id/101/800/600'},
    {"label": "Image 2", "image": 'https://picsum.photos/id/102/800/600'},
    {"label": "Image 3", "image": 'https://picsum.photos/id/103/800/600'}
]
    
gallery = IpySwiper(images, show_labels=True, thumbnail_fit="fill", container_max_width=800)
gallery  # This will display the gallery
```
